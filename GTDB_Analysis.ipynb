```
import os
import pandas as pd
import numpy as np

dataset="https://raw.githubusercontent.com/MirandaCJohnson/GlobalTerrorismData/master/GTDB-v03232019.csv"
table=pd.read_csv(dataset)
```

```
df=pd.DataFrame(data=table)

df.head(3)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

print("Total of White Extremist Terrorist attacks in the USA by Group from 1970-2017")
table['gname'].value_counts().plot(kind="pie", figsize=(6,18))
```

```
print('Frequency of Attack Methods')
table["nperps"].value_counts().plot(kind="bar")
```

```
# 03/24/2019
# ********* Pie chart of all white extremist terror attacks in the USA 1970-2017


labels=df['gname'].unique()
#print (labels)

#***  Find the total number of attacks
total_attacks=df['eventid'].count()
#print(total_attacks)

#***  Find total number of attacks per each extremist group.  
#***  (Divide by total_attacks)*100 for percentages to use in pie chart sizes
total_by_group=df.gname.value_counts()
#print(total_by_group)

we=(total_by_group[0]/total_attacks)*100
kkk=(total_by_group[1]/total_attacks)*100
nne=(total_by_group[2]/total_attacks)*100
ar=(total_by_group[3]/total_attacks)*100
wpp=(total_by_group[4]/total_attacks)*100

#print(we, kkk, nne, ar, wpp)

#***  Pie chart with labels and proportionally correct sections
sizes=[we, kkk, nne, ar, wpp]

colors=['#a58863', '#2996a7', '#d7cc88', '#adc835', '#3bc123']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors=colors, labels=labels, autopct='%.1f%%', startangle=90)

print("Pie Chart of Attacks by Group from 1970-2017")
ax1.axis('equal')
plt.tight_layout()
plt.legend(loc='upper left')
plt.show()
```
